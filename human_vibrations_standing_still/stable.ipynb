```python
import pandas as pd 
import numpy as np
from IPython.display import display
from scipy.signal import welch
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('VMD_EO', encoding='cp1251', sep='\s{2,}', engine='python') # Р·Р°РіСЂСѓР¶Р°РµРј РґР°С‚Р°СЃРµС‚d
df.head(2)
```

```python
df.info()
df.head(2)
print('index object:', df.index, '\n\ncolumns object:', df.columns, '\n\ndataframe shape:', df.shape) #СѓР·РЅР°РµРј С‚РёРїС‹ РґР°РЅРЅС‹С…
```

```python
# РџРµСЂРµРёРјРµРЅРѕРІР°РЅРёРµ СЃС‚РѕР»Р±С†РѕРІ СЃРѕРіР»Р°СЃРЅРѕ С„Р°Р№Р»Сѓ
df.columns = ['Stabilograph_X', 'Stabilograph_Y', 'Ballistogram', 'Breathing']
```

```python
# 2. РЎС‚Р°С‚РёСЃС‚РёС‡РµСЃРєРёРµ РїР°СЂР°РјРµС‚СЂС‹
sigma_x = np.std(df['Stabilograph_X'])
sigma_y = np.std(df['Stabilograph_Y'])

print(f'sigma_x = {sigma_x}')
print(f'sigma_y = {sigma_y}')
```

```python
# Р’С‹С‡РёСЃР»РµРЅРёРµ РєРѕРІР°СЂРёР°С†РёРѕРЅРЅРѕР№ РјР°С‚СЂРёС†С‹
Cov = np.cov(df['Stabilograph_X'], df['Stabilograph_Y'])
S = 2 * np.log(1/0.1) * np.sqrt(Cov[0, 0] * Cov[1, 1] - Cov[0, 1]**2)
print(f'S_el = {S}')
```

```python
# 2. РџР°СЂР°РјРµС‚СЂС‹ РґР»СЏ Welch
fs = 50  # С‡Р°СЃС‚РѕС‚Р° РґРёСЃРєСЂРµС‚РёР·Р°С†РёРё
window = 'hann'  # РѕРєРЅРѕ РҐР°РЅРЅР°
nperseg = 1024  # РґР»РёРЅР° СЃРµРіРјРµРЅС‚Р°
noverlap = 500  # РєРѕР»РёС‡РµСЃС‚РІРѕ РїРµСЂРµРєСЂС‹РІР°СЋС‰РёС…СЃСЏ РѕС‚СЃС‡РµС‚РѕРІ
nfft = 1024  # РєРѕР»РёС‡РµСЃС‚РІРѕ С‚РѕС‡РµРє РґР»СЏ FFT
```

```python
# Р¤СѓРЅРєС†РёСЏ РґР»СЏ СЂР°СЃС‡РµС‚Р° СЃРїРµРєС‚СЂР° РјРѕС‰РЅРѕСЃС‚Рё Рё РїРѕСЃС‚СЂРѕРµРЅРёСЏ РіСЂР°С„РёРєР°
def plot_welch(column, Xmax, Ymax, title):
    f, Pxx = welch(column, fs=fs, window=window, nperseg=nperseg, noverlap=noverlap, nfft=nfft)
    
    plt.figure()
    plt.semilogy(f, Pxx)
    plt.xlim([0, Xmax])
    plt.ylim([0, Ymax])
    plt.xlabel('Р§Р°СЃС‚РѕС‚Р° [Р“С†]')
    plt.ylabel('РЎРїРµРєС‚СЂР°Р»СЊРЅР°СЏ РїР»РѕС‚РЅРѕСЃС‚СЊ РјРѕС‰РЅРѕСЃС‚Рё [РєРі^2/Р“С†]')
    plt.title(title)
    plt.grid()
    plt.show()
```

```python
# 3. РџРѕСЃС‚СЂРѕРµРЅРёРµ РіСЂР°С„РёРєРѕРІ РґР»СЏ СЂР°Р·РЅС‹С… СЃС‚РѕР»Р±С†РѕРІ
# Р”Р»СЏ РєРѕРѕСЂРґРёРЅР°С‚ С†РµРЅС‚СЂР° РґР°РІР»РµРЅРёСЏ (Stabilograph_X Рё Stabilograph_Y): Xmax = 2 Р“С†
plot_welch(df['Stabilograph_X'], Xmax=2, Ymax=None, title='РЎРїРµРєС‚СЂ РјРѕС‰РЅРѕСЃС‚Рё РґР»СЏ Stabilograph_X')
plot_welch(df['Stabilograph_Y'], Xmax=2, Ymax=None, title='РЎРїРµРєС‚СЂ РјРѕС‰РЅРѕСЃС‚Рё РґР»СЏ Stabilograph_Y')

# Р”Р»СЏ РІРµСЃР° (Ballistogram): Xmax = 12 Р“С†
plot_welch(df['Ballistogram'], Xmax=12, Ymax=None, title='РЎРїРµРєС‚СЂ РјРѕС‰РЅРѕСЃС‚Рё РґР»СЏ Ballistogram')
```

```python
# Р Р°СЃС‡РµС‚ СЃРїРµРєС‚СЂР° РјРѕС‰РЅРѕСЃС‚Рё
Fx, Pxx = welch(df['Stabilograph_X'], fs=fs, window=window, nperseg=nperseg, noverlap=noverlap, nfft=nfft)
Fy, Pyy = welch(df['Stabilograph_Y'], fs=fs, window=window, nperseg=nperseg, noverlap=noverlap, nfft=nfft)
Fw, P = welch(df['Ballistogram'], fs=fs, window=window, nperseg=nperseg, noverlap=noverlap, nfft=nfft)

# 4. РџРѕСЃС‚СЂРѕРµРЅРёРµ РіСЂР°С„РёРєРѕРІ

# Р¤СЂРѕРЅС‚Р°Р»СЊРЅР°СЏ РѕСЃСЊ
plt.figure()
plt.plot(Fx, Pxx)
plt.xlim([0, 2])
plt.ylim([0, 6])
plt.xlabel('Р§Р°СЃС‚РѕС‚Р° [Р“С†]')
plt.ylabel('РЎРїРµРєС‚СЂР°Р»СЊРЅР°СЏ РїР»РѕС‚РЅРѕСЃС‚СЊ РјРѕС‰РЅРѕСЃС‚Рё [РєРі^2/Р“С†]')
plt.title('Р¤СЂРѕРЅС‚Р°Р»СЊ')
plt.grid()
plt.show()

# РЎР°РіРёС‚С‚Р°Р»СЊРЅР°СЏ РѕСЃСЊ
plt.figure()
plt.plot(Fy, Pyy)
plt.xlim([0, 2])
plt.ylim([0, 25])
plt.xlabel('Р§Р°СЃС‚РѕС‚Р° [Р“С†]')
plt.ylabel('РЎРїРµРєС‚СЂР°Р»СЊРЅР°СЏ РїР»РѕС‚РЅРѕСЃС‚СЊ РјРѕС‰РЅРѕСЃС‚Рё [РєРі^2/Р“С†]')
plt.title('РЎР°РіРёС‚Р°Р»СЊ')
plt.grid()
plt.show()

# Р‘Р°Р»Р»РёСЃС‚РѕРіСЂР°РјРјР°
plt.figure()
plt.plot(Fw, P)
plt.xlim([0, 12])
plt.ylim([0, 5e-3])
plt.xlabel('Р§Р°СЃС‚РѕС‚Р° [Р“С†]')
plt.ylabel('РЎРїРµРєС‚СЂР°Р»СЊРЅР°СЏ РїР»РѕС‚РЅРѕСЃС‚СЊ РјРѕС‰РЅРѕСЃС‚Рё [РєРі^2/Р“С†]')
plt.title('Р‘Р°Р»Р»РёСЃС‚РѕРіСЂР°РјРјР°')
plt.grid()
plt.show()

# 5. Р”РѕРїРѕР»РЅРёС‚РµР»СЊРЅС‹Рµ РіСЂР°С„РёРєРё
# РўСЂР°РµРєС‚РѕСЂРёСЏ (С„СЂРѕРЅС‚Р°Р»СЊ-СЃР°РіРёС‚Р°Р»СЊ)
plt.figure()
plt.plot(df['Stabilograph_X'], df['Stabilograph_Y'])
plt.xlabel('Р¤СЂРѕРЅС‚Р°Р»СЊ')
plt.ylabel('РЎР°РіРёС‚Р°Р»СЊ')
plt.title('РўСЂР°РµРєС‚РѕСЂРёСЏ РґРІРёР¶РµРЅРёСЏ')
plt.grid()
plt.show()

# Р“СЂР°С„РёРє Р±Р°Р»Р»РёСЃС‚РѕРіСЂР°РјРјС‹
# plt.figure()
# plt.plot(df['Ballistogram'])
# plt.xlabel('Р’СЂРµРјСЏ')
# plt.ylabel('Р‘Р°Р»Р»РёСЃС‚РѕРіСЂР°РјРјР°')
# plt.title('Р“СЂР°С„РёРє Р±Р°Р»Р»РёСЃС‚РѕРіСЂР°РјРјС‹')
# plt.grid()
# plt.show()
```
