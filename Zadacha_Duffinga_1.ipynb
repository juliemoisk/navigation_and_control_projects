```python
import numpy as np
import matplotlib.pyplot as plt

d = 5.0 # 0.0 5.0
g = 0.1 # +0.1 -0.1
w = 5.0 

def sys(q, p):                        # описываем систему, нормализуем значения
    dqdt = p                          # p = dq/dt
    dpdt = -2*d*p - w**2*(q + g*q**3) # dp/dt * 2*d*dq/dt + w^2*(q + g*q^3) = 0 ==> dp/dt = -2*d*p - w^2(q + g*q^3)
    #l = np.sqrt(dqdt**2 + dpdt**2)
    #dqdt = dqdt/l
    #dpdt = dpdt/l
    return np.array([dqdt, dpdt])
```

```python
q_v = np.linspace(-5, 5, 50)     # значения q
p_v = np.linspace(-10, 10, 100)  # значения p
Q, P = np.meshgrid(q_v, p_v)   # сетка начальных значений

# Вычисляем направления стрелок
dQ1, dQ2 = sys(Q, P)
```

```python
plt.streamplot(Q, P, dQ1, dQ2, color=np.sqrt(Q**2 + P**2), linewidth=1.5)
plt.title('Фазовая плоскость системы Дуффинга')
plt.xlabel('q')
plt.ylabel('p = dq/dt')
plt.xlim(-5, 5)
plt.ylim(-10, 10)
plt.axhline(0, color='k', lw=0.5, ls='--')
plt.axvline(0, color='k', lw=0.5, ls='--')
plt.grid()
plt.show()
```
