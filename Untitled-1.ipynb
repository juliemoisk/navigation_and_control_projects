```python
import numpy as np

def E_m(singular_values):
    # Вычисляем полную сумму сингулярных чисел
    total_sum = np.sum(singular_values)
    # Инициализация списка значений E_m
    E_values = []
    
    # Для каждого m (от 1 до n)
    for m in range(1, len(singular_values) + 1):
        # Сумма сингулярных значений от m+1 до n
        sum_after_m = np.sum(singular_values[m-1:])
        # Вычисление E_m
        E = sum_after_m / total_sum
        E_values.append(E)
    
    return E_values

def find_optimal_m(singular_values, threshold=0.2):
    # Вычисляем E_m для всех m
    E_values = E_m(singular_values)
    # Ищем минимальное m, для которого E_m < threshold
    for m, E in enumerate(E_values, start=1):
        if E < threshold:
            return m
    return len(singular_values)  # если не нашли, возвращаем максимальное m

# Загрузка данных
data = np.load('PCA.npy')  # загрузите свой файл с данными

# Выполнение SVD
_, s, _ = np.linalg.svd(data, full_matrices=False)

# Находим оптимальное значение m
optimal_m = find_optimal_m(s, threshold=0.2)
print("Оптимальное значение m:", optimal_m)
```
